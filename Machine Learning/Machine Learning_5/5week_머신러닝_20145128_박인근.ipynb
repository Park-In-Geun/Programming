## Maxtrix Multiplication

```python
import numpy as np
import pandas as pd
import os
import pickle


def makeMatrix():
    np.random.seed(20145128)
    matrix = np.floor(np.random.random(40) * 10)
    for i in range(40):
        if matrix[i] == 0:
            matrix[i] = np.NaN
    return matrix

def savefile(file_path,matrix):
    f = open(file_path,"wb")
    pickle.dump(matrix,f)
    f.close()
    
def loadfile(file_path):
    f = open(file_path,"rb")
    data = pickle.load(f)
    return data

def makeShape(matrix,row,col):
    mat = np.zeros((row,col))
    k = 0
    
    for i in range(0,row):
        for j in range(0,col):
            mat[i][j] = matrix[k]
            k = k + 1
            if k > 40:
                return 0
            
    return mat
        
def nanToZero(matrix):
    for i in range(40):
        matrix[np.isnan(matrix)] = 0
    return matrix
    
def matMul(mat1,mat2):
    matrix = np.zeros((8,8))
    
    for i in range(0,8):
        for j in range(0,8):
            for k in range(0,5):
                temp = mat1[i][k] * mat2[k][j]
                matrix[i][j] = matrix[i][j] + temp
            
    return matrix
    
def makePandas(matrix):
    datas = matrix
    rows = ["1행","2행","3행","4행","5행","6행","7행","8행"]
    columns = ["1열","2열","3열","4열","5열","6열","7열","8열"]
    df = pd.DataFrame(datas, index=rows,columns=columns)
    
    return df

if __name__ == "__main__":
    file_path = os.getcwd() + "/matrix.pkl"
    
    matrix = makeMatrix()
    print(matrix)
    savefile(file_path,matrix)
    
    del matrix
    matrix = loadfile(file_path)
    print(matrix)
    
    matrix = nanToZero(matrix)
    print(matrix)
    
    mat1 = makeShape(matrix,8,5)
    print(mat1)
    mat2 = makeShape(matrix,5,8)
    print(mat2)
    
    result = matMul(mat1,mat2)
    print(result)
    
    frame = makePandas(result)
    display(frame)
```
